```python
# implememting square and leaky relu in python
import numpy as np
def square(x: np.ndarray) -> np.ndarray:
    """
    Square each element in the array
    """
    return np.power(x, 2)

def leaky_relu(x: np.ndarray) -> np.ndarray:
    """
    Applying leaky relu
    """
    return np.maximum(0.2 * x, x)
```

```python
# Test square
arr_sq = np.array([-15, -10, -5, 0, 5, 10, 15, 20])
ex_sq = square(arr_sq)
print(ex_sq)

# Test leaky relu
arr_lrelu = np.array([-1, 3, 4, 0, -2, 9, 10, 5, -8, -7])
ex_lrelu = leaky_relu(arr_lrelu)
print(ex_lrelu)
```

```python
# Plotting
import matplotlib.pyplot as plt
%matplotlib inline

fig, axs = plt.subplots(1, 2, figsize=(9, 2), sharey=True)
axs[0].plot(arr_sq, ex_sq)
axs[1].plot(arr_lrelu, ex_lrelu)
fig.suptitle('Plotting Functions')

plt.plot(arr_sq, ex_sq, 'go--', linewidth=2, markersize=12)
```
